```python
from typing import List
Vector=List[float]
def add(v: Vector,w: Vector)->Vector:
    assert len(v)==len(w)
    return [vi+wi for vi,wi in zip(v,w)]
    assert add([1,2,3],[4,5,6])==[5,7,6]
```

```python
num_friends=[100,49,41,40,25,10,8,12,5,6,70,50,52]
from collections import  Counter
import  matplotlib.pyplot as plt
c=Counter(num_friends)
xs=range(101)
ys=[c[x] for x in xs]
plt.bar(xs,ys)
plt.axis([0,101,0,25])
plt.title("histrogram of friends count")
plt.xlabel("number of friends")
plt.ylabel("no of people")
plt.show()
```
